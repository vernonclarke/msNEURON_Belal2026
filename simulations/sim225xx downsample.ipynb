Run with **cell type** set to either **iSPN** or **dSPN**.

**Threshold** is defined as the number of clustered **glutamatergic inputs** required to cause an upstate.

In **dSPN** and **iSPN**, an upstate is observed with 15 and 14 clustered **glutamatergic inputs**, respectively.

In **50% K<sup>+</sup> reduction**, 14 and 13 clustered **glutamatergic inputs** remain subthreshold for **dSPN** and **iSPN**, respectively.

In **75% K<sup>+</sup> reduction**, 13 clustered **glutamatergic inputs** is suprathreshold for **iSPN** and is reduced to 12 for the simulation.
